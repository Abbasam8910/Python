# Python Dictionaries

```python
Programming_Dictionary={
    "Bug":"An error in a program", 
    "Function":"A Piece of code that easily call over and over again", 
    "Loop": "The Action to doing something over and over again"
    }


print(Programming_Dictionary,"\n")


# print(Programming_Dictionary["Bog"])  --Typing Error : KeyError: means this key does not exist
```

Retrieving items from dictionary

```python
# Retrieving items from dictionary
print(Programming_Dictionary["Loop"])
```

Create empty dictionary

```python
# Create empty dictionary
empty_dictionary={}
print(empty_dictionary)
```

Adding new item in dictionary

```python
# Adding new item in dictionary
Programming_Dictionary["New Item"]= "New item is added in dictionary"
print(Programming_Dictionary)
```

Wipe an existing dictionary

```python
# Wipe an existing dictionary
Programming_Dictionary={}

print(Programming_Dictionary)
```

Edit an item in dictionary

```python
Programming_Dictionary["Bug"]="New bug definition added"

print(Programming_Dictionary)
```

For Loop in Dictionary

```python
for key in Programming_Dictionary:
    print(key)

# Output Not comming as expected
```

```python
for key in Programming_Dictionary:
    print(key)
    print(Programming_Dictionary[key]) # Now you can print the key and description 
```
