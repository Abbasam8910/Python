# Day 8 Project: Caesar Cipher 

Encryption and Decryption using Ceasar Cipher

```python
alphabet=['a','b','c', 'd', 'e', 'f', 'g', 'h', 'i', 'j', 'k', 'l', 'm', 'n', 'o', 'p', 'q', 'r', 's', 't', 
'u', 'v', 'w', 'x', 'y', 'z','a','b','c', 'd', 'e', 'f', 'g', 'h', 'i', 'j', 'k', 'l', 'm', 'n', 'o', 'p', 
'q', 'r', 's', 't', 'u', 'v', 'w', 'x', 'y', 'z']




encrypt_decrypt=True

def encrypt(plain_text, shift_amount):
    cipher_text=""
    for letters in plain_text:
        position=alphabet.index(letters)
        New_position=position+shift_amount
        New_letter=alphabet[New_position]

        cipher_text=cipher_text+New_letter
    print(f"Encoded Text is {cipher_text}")   

def decrypt(plain_text, shift_amount):
    decrypt_text=""
    for letters in plain_text:
        position=alphabet.index(letters)
        New_position=position-shift_amount
        New_letter=alphabet[New_position]

        decrypt_text=decrypt_text+New_letter
    print(f"Decoded Text is {decrypt_text}")

while encrypt_decrypt==True:


    direction=input("Type 'encode' to encrypt, type 'decode' to decrypt").lower()
    text=input("Type you message: \n").lower()
    shift=int(input("Type a shift number: \n")) 

    if direction=="encode":
        encrypt(plain_text=text,shift_amount=shift)
    else:
        decrypt(plain_text=text,shift_amount=shift)
    
    try_again=input("Do you Want to Try Again: 'Yes' or 'No'").lower()

    if try_again=="yes":
        encrypt_decrypt=True
    else:
        encrypt_decrypt = False







```
