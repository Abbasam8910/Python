# List in Python


```python
states_of_India=["Kerala", "Karnataka", "Tamilnadu","Maharastra", "Gujarat","Assam"]

print(states_of_India[2])
print(states_of_India[-2])

states_of_India[1]="kerala"   # changing position in list
states_of_India[0]="karnataka"

print(states_of_India[0])
```

```python
# How Will Pay the Bill?

friends=["Ram", "Raj", "Kevin", "Alen", "Rahul", "Priya"]

# Option 1
print(random.choice(friends))  # Printing random from list.

# Option 2

random_index=random.randint(0,4)
print(friends[random_index])
```
