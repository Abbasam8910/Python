# Day 4 Project: Rock Paper Scissors

```python
import random
print("#################### WELCOME TO ROCK PAPER SCISSORS  GAME ##################")

choosen=int(input("What do you want to choose: \n Type 0 for Rock\n Type 1 for Paper \n Type 2 for Scissors"))

if choosen ==0:
    print("You have Choosen:  ")
    print("""
    _______
---'   ____)
      (_____)
      (_____)
      (____)
---.__(___)
""")
elif choosen ==1:
    print("You have Choosen:")
    print("""
     _______
---'    ____)____
           ______)
          _______)
         _______)
---.__________)
""")

else:
    print("You Have choosen: ")
    print("""
    _______
---'   ____)____
          ______)
       __________)
      (____)
---.__(___)
""")

choice_for_computer=["Rock", "Paper", "Scissors"]

choosen_computer=random.choice(choice_for_computer)

if choosen_computer =="Rock":
    print("Computer have Choosen:  ")
    print("""
    _______
---'   ____)
      (_____)
      (_____)
      (____)
---.__(___)
""")
elif choosen_computer =="Paper":
    print("Computer have Choosen:")
    print("""
     _______
---'    ____)____
           ______)
          _______)
         _______)
---.__________)
""")

else:
    print("Computer Have choosen: ")
    print("""
    _______
---'   ____)____
          ______)
       __________)
      (____)
---.__(___)
""")


if choosen==0 and choosen_computer=="Rock":
    print("Draw")
elif choosen==0 and choosen_computer=="Paper":
    print("You Failed")
elif choosen==0 and choosen_computer=="Scissors":
    print("You Win")

elif choosen==1 and choosen_computer=="Rock":
    print("You Win")
elif choosen==1 and choosen_computer=="Paper":
    print("Draw")
elif choosen==1 and choosen_computer=="Scissors":
    print("You Failed")

elif choosen==2 and choosen_computer=="Rock":
    print("You Failed")
elif choosen==2 and choosen_computer=="Paper":
    print("You Win")
elif choosen==2 and choosen_computer=="Scissors":
    print("Draw")
else:
    print("Some Errors")

```
